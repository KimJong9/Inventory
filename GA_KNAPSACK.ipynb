# 최대용량이 Y, 칸수가 X칸인 가방에 최대한 비싸게 담아보세요.

```python
import numpy as np
import math as mt
import random
import sys
```

초기 설정

```python
KNAPSACK = 2308               #배낭 용량, 단위 = g
gold = np.array([1,302,62])   #금 302g, 62 value
silver = np.array([2,200,41]) #은 200g, 41 value
dollar = np.array([3,99,20])  #달러 99g, 20value
empty = np.array([4,0,0])     #빈칸
```

```python
#기타 선언
num_gene = 20       # 유전자 갯수
num_inventory = 17   # 가방 최대 갯수
```

```python
#배열 선언
gene=np.zeros((num_gene,num_inventory),dtype=int)     #유전 배열
value = np.zeros((num_gene,num_inventory),dtype=int)  #가격 배열
weight = np.zeros((num_gene,num_inventory),dtype=int) #무게 배열
to_val = np.ndarray(num_gene,dtype=int) #total value of gene
pro = np.ndarray(num_gene,dtype=float)  #probability(value)
wei = np.ndarray(num_gene,dtype=int)    #total weight of gene
parent = np.zeros((2,num_inventory),dtype=int)
```

함수 선언

```python
#초기 유전자 생성 함수
def create_gene():
    for i in range(0,num_gene):
        for j in range(0,num_inventory):
            gene[i][j]=random.randint(1,4)
```

```python
#진화 함수 Evaluation_function
def evaluation(gene): 
    total_weight = 0    #total weight
    total_val = 0     #total value
    for i in range(0,num_gene):
        for j in range(0,num_inventory):
            if (gene[i][j]==gold[0]):
                weight[i][j]=gold[1]
                value[i][j]=gold[2]
            elif (gene[i][j]==silver[0]):
                weight[i][j]=silver[1]
                value[i][j]=silver[2]
            elif (gene[i][j]==dollar[0]):
                weight[i][j]=dollar[1]
                value[i][j]=dollar[2]
            elif (gene[i][j]==empty[0]):
                weight[i][j]=empty[1]
                value[i][j]=empty[2]
            else:
                print("오?류")
    for i in range(0,num_gene):
        for j in range(0,num_inventory):
            total_val = total_val + value[i][j]
            total_weight = total_weight + weight[i][j]
        wei[i]=total_weight
        to_val[i]=total_val
        total_weight = 0
        total_val = 0
```

```python
#각 유전자의 확률 구하기
def probability(pro):
    total = 0
    for i in range(0,num_gene):
        total = total+to_val[i]
    for i in range(0,num_gene):
        pro[i]=to_val[i]/total
```

```python
#생존시킬 부모 유전자 정하기 (단 무게가 초과하면 선택에서 제외)
def sel_parent(prob,weit):
    count = 0
    while(count!=2):
        for i in range (0,num_gene):
            if (prob[i]==max(prob) and max(prob)>0):   #최댓값을 찾아서
                if(weit[i]<=KNAPSACK): #무게 한계치보다 작을경우
                    if(count!=2):
                        parent[count]=gene[i]
                        count=count+1
                prob[i]=0
            if(max(prob)==0):
                break
```

```python
def mutation(gene):   #돌연변이는 해당 값을 공백으로 변경(이미 공백일경우 아무거나 채워넣기)
    for i in range(0,num_inventory):
        gene[i]=random.randint(1,4)
```

```python
def crossover(parent):
    temp = np.ndarray(num_inventory,dtype=int)
    gene[0]=parent[0]
    gene[1]=parent[1]
    for i in range(2,num_gene):
        standard=random.randint(1,num_inventory)
        for j in range(0,num_inventory):
            if(j<standard):
                gene[i][j]=gene[0][j]
            else:
                gene[i][j]=gene[1][j]
        o=random.randint(1,100)
        if (o<=50):
            mutation(gene[i])
```

```python
def calculation(count):
    tcount=0
    while (tcount<count):
        evaluation(gene)
        probability(pro)
        sel_parent(pro,wei)
        crossover(parent)
        tcount=tcount+1
    printmsg(tcount,gene[0],wei[0],to_val[0])
```

```python
def printmsg(count,gene,wei,to_val):
    gol=sil=cash = 0
    for i in range(num_inventory):
        if(gene[i]==1):
            gol=gol+1
        elif(gene[i]==2):
            sil=sil+1
        elif(gene[i]==3):
            cash=cash+1
    print("설정 - 유전자 개수:",num_gene,"개 가방용량:",KNAPSACK,"g 가방에 넣을 아이템 수:",num_inventory,"개")
    print(count,"번 계산 결과, 가방에 금 ",gol,"개, 은 ",sil,"개, 현금 ",cash,"다발을 넣는것을 추천합니다. 무게:",wei,"g, 가치:",to_val*10,"만원")
```
